```python
# Лабораторна робота №2
# студента гарупи АнД-31
# Пивовара Володимира
# Варіант №9

# Task 1 - За матеріалами ознайомитись із засадами бібліотеки TensorFlow:
# - засвоєння поняття тензору
# - знати різні способи створення тензорів
# - створити набір даних із заданим розподілом (розподіл та розмір за власним рішенням), визначити його властивості (shape, rank, size, axis)
# - продемонструвати основні математичні операції над тензорами
# - провести CRUD операції над даними, reshape, інтеграцію з np.array, pd.DataFrame)

import tensorflow as tf

tensor = tf.Variable([2, 2]) # Можна змінювати
tensor = tf.constant([2, 2]) # Не можна змінювати (константа)
tensor = tf.zeros([2, 2]) # Тензор заповнений нулями
tensor = tf.ones([2, 2]) # Тензор заповнений одиницями

# Створення випадкового тензору
tensor = tf.random.Generator.from_seed(42)
tensor = tensor.normal(shape=(7, 6))

# Визначення його властивостей
print(tensor)
print("\nTensor shape: ", tensor.shape)
print("Tensor size:", tf.size(tensor).numpy())
print("Tensor rank(.ndim()):", tensor.ndim)
print("Tensor elements on \'0\' axis:", tensor.shape[0])
print("Tensor elements on \'1\' axis:", tensor.shape[1])

# Основні математичні операції
print('\nOPERATIONS TENSOR / NUMBER\nAddition:\n', tensor + 1)
print('\nSubtraction:\n', tensor - 2)
print('\nMultiplication:\n', tensor * 1.5)
print('\nDivision:\n', tensor / 2)
print('\nOPERATION TENSOR / TENSOR')
tensor2 = tf.random.Generator.from_seed(40)
tensor2 = tensor2.normal(shape=(6, 3))
print('First tensor:\n', tensor)
print('Second tensor:\n', tensor2)
print('\nMatrix mutliplication\n', tensor @ tensor2)
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd

tensor = tf.random.Generator.from_seed(42)
tensor = tensor.normal(shape=(3, 4))

print(tensor)
print('\nTransposition:\n', tf.transpose(tensor))
print('\nReshape:\n', tf.reshape(tensor, shape=(2, 6)))

tensor = tf.constant(np.zeros([3, 4]))
print('\nCreate tensor with numpy:\n', tensor)
arr = np.array(tensor)
print('\nConvert tensor to numpy:\n', arr)
tensor = tf.Variable(arr)
print('\nConvert numpy to tensor:\n', tensor)

numbers = pd.DataFrame({
    '0': [36, 33, 29],
    '1': [31, 30, 33],
    '2': [21, 18, 17],
    '3': [14, 32, 28]
})
print('\nDataFrame:\n', numbers)
tensor = tf.Variable(numbers)
print('\nConvert to tensor:\n', tensor)
```
