```python
# Лабораторна робота №1
# студента гарупи АнД-31
# Пивовара Володимира
# Варіант №9

# Task 2 - NumPy (створити набір даних (розмір за власним рішенням), провести CRUD операції над даними (скаляр, вектор, матриця)). Властивість shape та метод  reshape. Операції над
# масивами. Використати arange, random.

import numpy as np

# CRUD операції над вектором
vector = np.arange(7) # Створення
for i in range(len(vector)):
  vector[i] = np.random.randint(0, 100) # Зміна
print('Vector:', vector)
vector = np.delete(vector, 1) # Видалення
print('After delete:', vector)

# CRUD операції над матрицею
arr = np.array([[0] * 7] * 8) # Створення
for i in range(len(arr)):
  arr[i] = np.random.randint(0, 100, 7) # Зміна
print('\nArray:', arr)
arr = np.delete(arr, 4, 1) # Видалення
print('After delete:\n', arr)
```

```python
# Task 3 - Pandas (створити dataframe різними способами, провести маніпуляції з даними (head, describe, iloc, loc, індексація з умовами відбору, групування,
# пошук «пустих» значень, сортування)

import pandas as pd

teams = pd.DataFrame({
    'Назва': ['Динамо', 'Шахтар', 'Боруссія', 'Баварія', 'Ювентус', 'Лаціо'],
    'Країна': ['Україна', 'Україна', 'Німеччина', 'Німеччина', 'Італія', 'Італія'],
    'Місто': ['Київ', 'Донецьк', 'Дортмунд', 'Мюнхен', 'Турин', 'Рим'],
    'К-ть гравців': [36, 33, 29, 31, 30, 33]
})
print('Перший варіант створення:\n', teams)

name = ['Динамо', 'Шахтар', 'Боруссія', 'Баварія', 'Ювентус', 'Лаціо']
country = ['Україна', 'Україна', 'Німеччина', 'Німеччина', 'Італія', 'Італія']
city = ['Київ', 'Донецьк', 'Дортмунд', 'Мюнхен', 'Турин', 'Рим']
players = [36, 33, 29, 31, 30, 33]
teams = pd.DataFrame({
    'Назва': name,
    'Країна': country,
    'Місто': city,
    'К-ть гравців': players
})
print('\nДругий варіант створення:\n', teams)

print('\n\'teams.head\' - Перші три значення:\n', teams.head(3))
print('\n\'teams.describe - Описова статистика\':\n', teams.describe())
print('\n\'iloc\' - Обрати певний елемент:\n', teams.iloc[0])
print('\n\'loc\' - Обрати елементи за умовою:\n', teams.loc[teams['Країна'] == 'Італія'])
print('\nІндексація з умовами відбору (назва і місто команди, де більше або рівно 30 гравців):\n', teams.loc[teams['К-ть гравців'] >= 30][['Назва', 'Місто']])
print('\nГрупування за країною та визначення середньої кількості гравців:\n', teams.groupby('Країна').mean())
teams.at[2, 'Місто'] = None
teams.at[5, 'Країна'] = None
print('\nЗмінений dataframe:\n', teams)
print('\nПошук пустих значень:\n', teams.isnull())
print('\nСортування (за алфавітним порядком):\n', teams.sort_values('Назва'))
```

```python
# Task 4 - Matplotlib (створити набір даних лінійної функції з урахуванням помилок вимірювань, побудувати графік функції та згенерованих даних, обчислити похибку з
# використанням метрик MAE, MSE). Записати отримані результати у csv файл (формат: X,Y, Y_hat, mAE, mSE)

import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
import numpy as np
import pandas as pd

# Функція: y = 3 * x + 1
X = [-3, -2, -1, 0, 1, 2, 3, 4]
Y = [-7.7, -5, -1.8, 1.3, 4.2, 6.6, 10.5, 13.1]

Y_hat = [0] * len(Y)
for i in range(len(Y)):
  Y_hat[i] = 3 * X[i] + 1

plt.scatter(X, Y, c = 'black')
plt.plot(X, Y_hat)

mAE = np.sqrt(mse(Y, Y_hat))
mSE = mae(Y, Y_hat)
print('\nmAE =', mAE, '\nmSE =', mSE)

result = pd.DataFrame({
    'X': X,
    'Y': Y,
    'Y_hat': Y_hat,
    'mAE': mAE,
    'mSE': mSE
})

result.to_csv('resultMatplotlib.csv')
```
